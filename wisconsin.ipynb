```python
import numpy as np  # linear algebra
import pandas as pd  # data processing

from sklearn.tree import DecisionTreeRegressor  # model

from matplotlib import pyplot as plt  # visualization
from sklearn.tree import plot_tree  # visualize tree
```

```python
df_wisconsin = pd.read_csv('/content/sample_data/wisconsin.csv')
df_wisconsin
#print(df_loan.dtypes)
```

```python
df_wisconsin = df_wisconsin.loc[df_wisconsin.index.isin([i for i in range(1000)])]
```

```python
model = DecisionTreeRegressor(
    criterion='squared_error',
    splitter='best',
    min_samples_split=4,
    max_depth = 5 
)

# CART algorithm that builds the tree based on our data
model.fit(
    X=df_wisconsin.drop(columns=['Class']), 
    y=df_wisconsin['Class']
)
```

```python
plt.figure(figsize=(30, 50))
plot_tree(
    model,
    feature_names=['ClumpThickness','CellSize','CellShape','MarginalAdhesion','EpithelialSize','BareNuclei','BlandChromatin','NormalNucleoli','Mitoses'],
    impurity=False,
    filled=True,
    fontsize=18
)
plt.show()
```
